```python
import pandas as pd
import numpy as np
from astropy.utils.console import ProgressBar

from tqdm import tqdm_notebook as tqdm

from matplotlib import pyplot as plt
%matplotlib inline
```

```python
df_selected = pd.read_hdf('media/selected.hdf5', key='selection3')
```

```python
df_selected = df_selected.sort_values(by=['mass_bin', 'dnf_bin'])
```

```python
row_template = (
"""
<tr>
  <th scope="row">{num:d}</th>
  <td>{rockstarID:d}</td>
  <td>{mvir:.1e}</td>
  <td>{rvir:.1f}</td>
  <td>{dnf:.1f}</td>
  <td>{massbin:d}</td>
  <td>{dnfbin:d}</td>
</tr>
"""
)
```

```python
text = ''

for i in range(len(df_selected)):
    row = df_selected.iloc[i]
    
    row_text = row_template.format(
        num=i+1,
        rockstarID=row.name,
        rvir=row.rvir,
        mvir=row.mvir,
        dnf=row.dnf,
        massbin=int(row.mass_bin),
        dnfbin=int(row.dnf_bin),
    )
    
    text += row_text
    
print(text)
```
